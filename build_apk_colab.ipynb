# FitTracker APK Builder

One-click APK build for Android.

## Usage
1. **Runtime → Run all**
2. Wait ~15-30 min (first time downloads SDK/NDK)
3. APK downloads automatically at the end

```python
# Step 1: Install build tools
print("Installing Buildozer...")

!apt-get update -qq
!apt-get install -y -qq python3-pip python3-dev \
    libsdl2-dev libsdl2-image-dev libsdl2-mixer-dev libsdl2-ttf-dev \
    libportmidi-dev libswscale-dev libavformat-dev libavcodec-dev \
    zlib1g-dev libgstreamer1.0-0 gstreamer1.0-plugins-base \
    pkg-config autoconf automake libtool libltdl-dev unzip zip 2>&1 | tail -5

!pip install -q buildozer cython 2>&1 | tail -3

print("Build tools ready")
```

```python
# Step 2: Upload your files
# Upload main.py, database.py, buildozer.spec to /content/fittracker/
import os
os.makedirs('/content/fittracker', exist_ok=True)

# If you have a zip file, uncomment and run:
# !unzip /content/fittracker.zip -d /content/fittracker/

print("Files ready. Make sure main.py, database.py, buildozer.spec are in /content/fittracker/")
!ls -la /content/fittracker/
```

```python
# Step 3: Verify buildozer.spec
import os
print("buildozer.spec exists:", os.path.isfile("/content/fittracker/buildozer.spec"))
print()
!head -30 /content/fittracker/buildozer.spec
```

```python
# Step 4: Build APK
import time
start = time.time()

print("=" * 50)
print("Building APK...")
print("First build: ~15-30 min (SDK/NDK download)")
print("=" * 50)

!cd /content/fittracker && yes | buildozer android debug 2>&1

elapsed = time.time() - start
print(f"Build complete! Elapsed: {elapsed/60:.1f} min")
```

```python
# Step 5: Download APK
from google.colab import files
import glob, os

apks = glob.glob("/content/fittracker/bin/*.apk")
if apks:
    for apk in apks:
        size = os.path.getsize(apk) / 1024 / 1024
        print(f"APK: {apk} ({size:.1f} MB)")
        files.download(apk)
else:
    print("No APK found. Check build logs above.")
```
